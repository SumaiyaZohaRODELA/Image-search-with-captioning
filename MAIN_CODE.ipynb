````python
import os
import torch
import faiss
import numpy as np
from PIL import Image
from transformers import CLIPProcessor, CLIPModel
import matplotlib.pyplot as plt
````

load clip model

````python
device = "cuda" if torch.cuda.is_available() else "cpu"

model = CLIPModel.from_pretrained("openai/clip-vit-base-patch32").to(device)
processor = CLIPProcessor.from_pretrained("openai/clip-vit-base-patch32")
````

````python
from google.colab import drive
import os
import pandas as pd

# Mount Google Drive
drive.mount('/content/drive')

# Path to your CSV file
csv_path = "/content/drive/My Drive/animal.csv"

# Check if file exists
if os.path.exists(csv_path):
    df = pd.read_csv(csv_path)
    print(f"✅ Loaded 'animal.csv'")
    print(f"   Shape: {df.shape}")
    print(f"   Columns: {list(df.columns)}")
    print("\n--- First 5 rows ---")
    print(df.head())
else:
    print(f"❌ File not found at: {csv_path}")
    print("\nListing contents of My Drive:")
    for f in os.listdir("/content/drive/My Drive"):
        print(f"  - {f}")
````

Convert Images to Embeddings

````python
image_paths = [os.path.join("/content/drive/My Drive", path) for path in df['image_path'].tolist()]
print(f"Found {len(image_paths)} image paths with full paths.")
````

````python
image_embeddings = []

for path in image_paths:
    # Construct the full path explicitly, assuming 'path' might still be relative
    full_image_path = os.path.join("/content/drive/My Drive", path)
    image = Image.open(full_image_path).convert("RGB")

    inputs = processor(images=image, return_tensors="pt").to(device)

    with torch.no_grad():
        output = model.get_image_features(**inputs)

    # Access the pooler_output from the BaseModelOutputWithPooling object
    embedding = output.pooler_output.cpu().numpy()

    # Normalize
    embedding = embedding / np.linalg.norm(embedding)

    image_embeddings.append(embedding)

image_embeddings = np.vstack(image_embeddings)
````

Create FAISS Index

FAISS stores vectors efficiently.

````python
dimension = image_embeddings.shape[1]

index = faiss.IndexFlatIP(dimension)

index.add(image_embeddings)

print("Total images indexed:", index.ntotal)
````

Search with Text Query

````python
query = "A scary cat"

inputs = processor(text=query, return_tensors="pt").to(device)

with torch.no_grad():
    output = model.get_text_features(**inputs)

text_embedding = output.pooler_output.cpu().numpy()

# Normalize
text_embedding = text_embedding / np.linalg.norm(text_embedding)
````

Retrieve Similar Images

````python
top_k = 3

scores, indices = index.search(text_embedding, top_k)

print("Top matches:")
````

Show Retrieved Images

````python
fig, axes = plt.subplots(1, top_k, figsize=(15, 5))
axes = axes.flatten() # Ensure axes is iterable even if top_k is 1

for i, idx in enumerate(indices[0]):
    image_path = image_paths[idx]

    print(f"Attempting to display image {i+1} (index {idx}) from path: {image_path}")

    if not os.path.exists(image_path):
        print(f"ERROR: Image file not found at: {image_path}. Skipping this image.\n")
        axes[i].set_title(f"Rank {i+1}: Not Found")
        axes[i].axis("off")
        continue

    try:
        img = Image.open(image_path).convert("RGB")
        # Print metadata to confirm image loading
        print(f"  Image properties: Size={img.size}, Format={img.format}, Mode={img.mode}")

        axes[i].imshow(img)
        axes[i].set_title(f"Rank {i+1}: {os.path.basename(image_path)}")
        axes[i].axis("off")
        print("Successfully processed and attempted to display image.\n")
    except Exception as e:
        print(f"ERROR: Could not open or display image {image_path}: {e}. Skipping this image.\n")
        axes[i].set_title(f"Rank {i+1}: Error")
        axes[i].axis("off")

plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.show()
````

Now, let's load a multimodal model (like `gemini-pro-vision`) which can take both text prompts and images as input. We'll use it to generate a caption for one of the retrieved images.

````python
# Import necessary libraries for Hugging Face BLIP
from transformers import BlipProcessor, BlipForConditionalGeneration
from PIL import Image
import torch

# Load the BLIP model and processor
# Move to the device where the CLIP model is running ('cpu' or 'cuda')
processor_blip = BlipProcessor.from_pretrained("Salesforce/blip-image-captioning-base")
model_blip = BlipForConditionalGeneration.from_pretrained("Salesforce/blip-image-captioning-base").to(device)

# Pick one of the retrieved images to caption (e.g., the first one)
# Ensure image_paths and indices are defined from previous cells
sample_image_path = image_paths[indices[0][0]]
print(f"Generating caption for image: {sample_image_path}")

# Load the image using PIL
img_to_caption = Image.open(sample_image_path).convert("RGB")

# Process the image for the model and move to the same device
inputs_blip = processor_blip(img_to_caption, return_tensors="pt").to(device)

# Generate a caption
out = model_blip.generate(**inputs_blip, max_new_tokens=50)

# Decode the generated tokens to text
caption = processor_blip.decode(out[0], skip_special_tokens=True)

print("\n--- Generated Caption (Hugging Face BLIP) ---")
print(caption)
print("---------------------------------------------")

# You can now use this caption for further analysis or comparison with your query.
````
